```python
# Cell 4 (Upgraded): Advanced Risk Mapping using FUND NAME + CATEGORY

def map_risk_advanced(name, category):
    n = str(name).lower() if pd.notna(name) else ""
    c = str(category).lower() if pd.notna(category) else ""

    text = n + " " + c  # combine both for matching

    # ------------------ HIGH RISK ------------------
    high_keywords = [
        "small cap", "smal cap", "micro cap",
        "thematic", "sector", "digital", "technology",
        "pharma", "healthcare", "defense", "infrastructure",
        "manufacturing", "banking", "psu",
        "international", "global", "overseas",
        "opportunities", "quant", "fof", "fund of funds"
    ]
    for kw in high_keywords:
        if kw in text:
            return "High"

    # ------------------ MODERATE RISK ------------------
    moderate_keywords = [
        "mid cap", "midcap", "multi cap", "multicap",
        "flexi cap", "flexicap", "elss", "tax saver",
        "contra", "value fund", "focused", "diversified",
        "large & mid", "large and mid"
    ]
    for kw in moderate_keywords:
        if kw in text:
            return "Moderate"

    # ------------------ LOW RISK ------------------
    low_keywords = [
        "large cap", "bluechip", "nifty", "sensex",
        "index fund", "etf", "exchange traded",
        "hybrid conservative", "conservative",
        "debt", "liquid", "overnight", "money market",
        "corporate bond", "short duration", "medium duration",
        "gilt", "psu bond", "credit risk"
    ]
    for kw in low_keywords:
        if kw in text:
            return "Low"

    # ------------------ FALLBACK ------------------
    return "Moderate"


# Apply the new risk mapping
df["Mapped_Risk"] = df.apply(
    lambda row: map_risk_advanced(row["Fund Name"], row["Category"]),
    axis=1
)

df["Mapped_Risk"].value_counts()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression





# Create composite target
df["Composite"] = (
    0.5*df["3Y Return"] +
    0.3*df["1Y Return"] +
    0.2*df["5Y Return"]
)

targets = {
    "1Y_target": "1Y Return",
    "3Y_target": "3Y Return",
    "5Y_target": "5Y Return",
    "AUM_target": "AUM",
    "Expense_target": "Expense Ratio",
    "Composite_target": "Composite"
}

def compute_weights(target_col):
    y = df[target_col].values
    X = df[['5Y Return','3Y Return','1Y Return','AUM','Expense Ratio']].values

    # Correlation weights
    corr = []
    for col in ['5Y Return','3Y Return','1Y Return','AUM','Expense Ratio']:
        corr.append(abs(df[col].corr(df[target_col])))
    corr = np.array(corr)
    corr_w = corr / corr.sum() if corr.sum()!=0 else np.zeros_like(corr)

    # Regression weights
    reg = LinearRegression()
    reg.fit(X, y)
    coef = np.abs(reg.coef_)
    reg_w = coef / coef.sum() if coef.sum()!=0 else np.zeros_like(coef)

    final = (corr_w + reg_w) / 2
    return dict(zip(["5Y","3Y","1Y","AUM","Expense"], final))

all_weights = {}
for name, col in targets.items():
    all_weights[name] = compute_weights(col)

pd.DataFrame(all_weights).T
```

```python
# ============================
# CELL 6: FINAL SCORING ENGINE
# Using Composite-Target Optimal Weights
# ============================

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Features to normalize
pos_feats = ['5Y Return', '3Y Return', '1Y Return', 'AUM']  # higher = better
neg_feats = ['Expense Ratio']                               # lower = better

# Normalize positive features
df_pos = pd.DataFrame(
    scaler.fit_transform(df[pos_feats]),
    columns=[f"norm_{c.replace(' ','_')}" for c in pos_feats]
)

# Normalize negative features and invert them
df_neg = pd.DataFrame(
    scaler.fit_transform(df[neg_feats]),
    columns=[f"norm_{c.replace(' ','_')}" for c in neg_feats]
)
df_neg = 1 - df_neg  # invert (lower expense ratio = higher score)

# Combine normalized features with df
df_norm = pd.concat(
    [df.reset_index(drop=True), df_pos.reset_index(drop=True), df_neg.reset_index(drop=True)],
    axis=1
)

# ============================
# FINAL OPTIMAL WEIGHTS
# ============================
weights = {
    'norm_5Y_Return'     : 0.218949,
    'norm_3Y_Return'     : 0.413275,
    'norm_1Y_Return'     : 0.270023,
    'norm_AUM'           : 0.039195,
    'norm_Expense_Ratio' : 0.058557
}

# Compute weighted score
df_norm['fund_score_raw'] = (
    df_norm['norm_5Y_Return']     * weights['norm_5Y_Return'] +
    df_norm['norm_3Y_Return']     * weights['norm_3Y_Return'] +
    df_norm['norm_1Y_Return']     * weights['norm_1Y_Return'] +
    df_norm['norm_AUM']           * weights['norm_AUM'] +
    df_norm['norm_Expense_Ratio'] * weights['norm_Expense_Ratio']
)

# Scale to 0–100
min_raw = df_norm['fund_score_raw'].min()
max_raw = df_norm['fund_score_raw'].max()
df_norm['fund_score_100'] = 100 * (df_norm['fund_score_raw'] - min_raw) / (max_raw - min_raw)

# Final Top 10 Funds
df_norm.sort_values('fund_score_100', ascending=False).head(10)
```

```python
# ==========================================
# CELL 7: SIP RECOMMENDATION FUNCTION
# ==========================================

import numpy as np

# Expected CAGR based on risk
risk_expected_return = {
    "Low": 0.105,       # 10.5%
    "Moderate": 0.13,   # 13%
    "High": 0.165       # 16.5%
}

# SIP with Step-up Calculator
def stepup_sip_value(monthly_sip, years, stepup_percent, expected_return):
    r = expected_return / 12  # monthly rate
    total_value = 0
    sip_amount = monthly_sip

    for yr in range(years):
        for m in range(12):
            total_value += sip_amount * ((1 + r)**((years*12) - (yr*12 + m)))
        sip_amount *= (1 + stepup_percent / 100)

    return total_value


# RECOMMENDATION FUNCTION
def recommend_funds(df_norm, risk, monthly_sip, stepup_percent, years, top_n=5):

    # 1. Pick expected return based on risk level
    expected_return = risk_expected_return[risk]

    # 2. Filter based on risk
    df_filtered = df_norm[df_norm["Mapped_Risk"] == risk]

    # 3. Sort by final composite score
    df_top = df_filtered.sort_values("fund_score_100", ascending=False).head(top_n)

    # 4. Compute future SIP value for user
    fv = stepup_sip_value(monthly_sip, years, stepup_percent, expected_return)

    # 5. Format result
    return df_top[["Fund Name", "Category", "AUM", "Expense Ratio", "fund_score_100"]], fv


# ==========================================
# Example run
# ==========================================

example_top_funds, example_future_value = recommend_funds(
    df_norm,
    risk="Moderate",
    monthly_sip=5000,
    stepup_percent=10,
    years=10
)

example_top_funds, example_future_value
```
